```
#check the column names if they are simmilar
dt1_colname=Telcom_dataset1.columns
dt2_colname=Telcom_dataset2.columns
dt3_colname=Telcom_dataset3.columns

print(dt1_colname)
print(dt2_colname)
print(dt3_colname)
```

```
#Rename the column names to be uniform on the data sets.

Telcom_dataset1.columns =['PRODUCT','VALUE', 'DATE_TIME', 'CELL_ON_SITE', 'DW_A_NUMBER','DW_B_NUMBER', 'COUNTRY_A', 'COUNTRY_B', 'CELL_ID', 'SITE_ID']
Telcom_dataset2.columns =['PRODUCT','VALUE', 'DATE_TIME', 'CELL_ON_SITE', 'DW_A_NUMBER','DW_B_NUMBER', 'COUNTRY_A', 'COUNTRY_B', 'CELL_ID', 'SITE_ID']
Telcom_dataset3.columns =['PRODUCT','VALUE', 'DATE_TIME', 'CELL_ON_SITE', 'DW_A_NUMBER','DW_B_NUMBER', 'COUNTRY_A', 'COUNTRY_B', 'CELL_ID', 'SITE_ID']
dt1_colname=Telcom_dataset1.columns
dt2_colname=Telcom_dataset2.columns
dt3_colname=Telcom_dataset3.columns
#Printout the results to check if the changes have been effected.
print(dt1_colname)
print(dt2_colname)
print(dt3_colname)
```

```
#Appending the three simillar data sets

Telcom_df_temp =Telcom_dataset1.append(Telcom_dataset2)
Telcom_df_temp

Telcom_df=Telcom_df_temp.append(Telcom_dataset3)

#preview of the shape of DataFrame
Telcom_df.shape
```

```
#Drop rows with NaN Values in DataFrame

Telcom_df= Telcom_df.dropna()
Telcom_df

#check if there are any missing values
#The search for missings is usually one of the first steps in data analysis. 
#At the beginning, the question is whether there are any missings at all and, if so, how many there are
Telcom_df.isna().any().any()
```

```
#The frequency of Missings values
Telcom_df.isna().sum().sum()
```

```
# dropping duplicate values 
Telcom_df.drop_duplicates(keep=False,inplace=True) 

#Find Duplicate Rows based on all columns on Telcom_df data frame
Telcom_df.duplicated()
```

```
#Joining the two Data sets Telcom_df and Cells_geo data set
#SITE_CODE= geo
#SITE_ID

result_df = pd.merge(Telcom_df,
                 cells_geo[['VILLES','STATUS','LOCALISATION','DECOUPZONE','ZONENAME','REGION','AREA','CELL_ID','SITE_CODE' ]],
                            left_on='SITE_ID',right_on='SITE_CODE',how='inner')
result_df
```

```
#Rename column VILLES to City

result_df= result_df.rename(columns={"VILLES": "City"})
result_df.head()
```

```
## Convert date from string to date times
#result_df['DATE_TIME'] = result_df['DATE_TIME'].apply(dateutil.parser.parse, dayfirst=True)

result_df['DATE_TIME'] = pd.to_datetime(result_df['DATE_TIME'])

result_df['DATE_TIME'] = result_df['DATE_TIME']

result_df
```

# New Section

```
#Which were the busiest cities for the three days?
Bussiest_cities = result_df.groupby([result_df['DATE_TIME'].dt.day,'City'])["VALUE"].sum().sort_values(ascending=False)

Bussiest_cities= pd.DataFrame(Bussiest_cities)

Bussiest_cities =Bussiest_cities.groupby('DATE_TIME').head(3)
Bussiest_cities
```

```
#compare product Income
product_Income= pd.DataFrame(result_df.groupby(result_df['PRODUCT'])["VALUE"].sum().sort_values(ascending=False))
product_Income
```

```
#compare product Popularity
product_popularity= pd.DataFrame(result_df.groupby(result_df['PRODUCT'])["VALUE"].count().sort_values(ascending=False))
product_popularity
```

```
#Product uptake per city

uptake_per_city=pd.DataFrame( result_df.groupby([result_df['City'],'PRODUCT'])["VALUE"].count().sort_values(ascending=False))

uptake_per_city

```
